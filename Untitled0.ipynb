#Detecção de faces com Haar Ccascade

```python
imagem = cv2.imread('/person.jpg')
```

```python
imagem.shape
```

```python
tamanhoImagemRGB = 500*560*3
tamanhoImagemRGB
```

```python
cv2_imshow(imagem)
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
imagem_cinza.shape
```

```python
cv2_imshow(imagem_cinza)
```

```python
detector_facial = cv2.CascadeClassifier('/haarcascade_frontalface_default.xml')
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza)
```

```python
deteccoes
```

```python
len(deteccoes)
```

```python
for(x, y, w, h) in deteccoes:
  print(x,y, w, h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 255), 3)
cv2_imshow(imagem)
```
